  Pada *dataset* ini adalah kumpulan data performa iklan secara global  dari platform google, meta, dan tiktok. Berdasarkan
tujuannya platform iklan yang menguntungkan didasari oleh *ROAS* untuk meningkatkan keuntungan dan anggaran pemasarannya menjadi efisien karena setiap platform mengeluarkan biaya dan menghasilkan pendapatan yang berbeda-beda. Perusahaan iklan ingin mengetahui iklan di platform mana yang paling efektif meningkatkan keuntungan.
